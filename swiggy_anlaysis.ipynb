```python
data_Highest_Rating_Restaurants = df[df['Rating'] >= 4.0]
data_Highest_Rating_Restaurants
```

```python
data_Highest_Rating_Restaurants = data_Highest_Rating_Restaurants.loc[:, ["Shop_Name", "Rating", "Cost_for_Two"]]
data_Highest_Rating_Restaurants
```

```python
data_Highest_Rating_Restaurants = data_Highest_Rating_Restaurants.groupby(["Shop_Name", "Rating"])["Cost_for_Two"].agg("mean")
data_Highest_Rating_Restaurants = data_Highest_Rating_Restaurants.reset_index()
data_Highest_Rating_Restaurants
```
